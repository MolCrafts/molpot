```python
import torch
import molpot as mpot
from molpot import alias
```

```python
# create lj system
n_atoms = 128
box = torch.tensor([10.0, 10.0, 10.0])
xyz = torch.rand((128, 3)) * box
frame = mpot.Frame()
frame[alias.xyz] = xyz
frame[alias.cell] = box
frame[alias.pbc] = torch.tensor([True, True, True])
```

```python
lj126 = mpot.PotentialSeq(mpot.potential.classic.LJ126())
```

```python
from molpot.engine.md.neighborlist import NeighborList
from molpot.engine.md.integrator import VelocityVerlet


md = mpot.engine.MoleculeDymanics(grad_required=False, work_dir="md")
md.add_potential(lj126)
md.add_module(
    VelocityVerlet(1.0)
)
md.add_module(
    NeighborList(2.5, 0.5)
)
# md.add_module(
#     ThermoOutput(100)
# )
# md.add_module(
#     XYZOutput(100)
# )
md.run(frame, 1000)
```
