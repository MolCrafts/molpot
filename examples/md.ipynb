```python
%load_ext autoreload
%autoreload 2
import torch
import molpot as mpot
from molpot import alias
```

```python
# create lj system
n_atoms = 128
box = torch.tensor([10.0, 10.0, 10.0])
xyz = torch.rand((128, 3)) * box
frame = mpot.Frame()
frame[alias.atom_types] = torch.zeros(n_atoms, dtype=torch.int64)
frame[alias.xyz] = xyz
frame[alias.cell] = box
frame[alias.atom_masses] = torch.ones((n_atoms, 1))
```

```python
lj126 = mpot.potential.classic.LJ126()
energy_reducer = mpot.Reducer("energy", "lj126_energy")
force_reducer = mpot.Reducer("force", "lj126_force")
potential = mpot.PotentialSeq(lj126, energy_reducer, force_reducer)
```

```python
from molpot.engine.md.events import MDMainEvents
from molpot.engine.md.neighborlist import NeighborList
from molpot.engine.md.integrator import VelocityVerlet
from molpot.engine.md.output import ThermoOutput, Thermo

md = mpot.engine.MoleculeDymanics(grad_required=False, work_dir="md")
md.set_potential(potential)
md.add_handlers(
    VelocityVerlet(1.0),
    NeighborList(2.5, 0.5),
    ThermoOutput(100, Thermo("step"), print),
)
md.run(frame, 100)
```
